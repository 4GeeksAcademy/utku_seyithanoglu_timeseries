## ARIMA

"ARIMA" stands for
* A.uto R.egression
* I.ntegrated
* M.oving A.verage

```python
sns.lineplot(x['T'].diff())
```

```python
sns.lineplot(copy_df['T'])
plt.axvline(pd.to_datetime('2005-03-01'),color='Black')
plt.grid(alpha=0.6)
```

```python
## Let's build a predictive model (on the Temperature variable)
```

```python
cutoff_date= pd.to_datetime('2005-03-01')

time_df=copy_df['T']

training_set = time_df[time_df.index < cutoff_date]
testing_set = time_df[time_df.index >= cutoff_date]




# “Model, 1 Mart 2005’e kadar olan verilerle eğitildi ve bu tarihten sonraki değerler üzerinde test edildi.”
```

```python
sns.lineplot(data=training_set)
sns.lineplot(data=testing_set)
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
model = ARIMA(endog=training_set,order=(2,  # p the number of autoregression terms  p = kaç gün geriye bakıyorum
                                         1, # d the number of differencing terms d, “kaçtayım?” yerine “ne kadar değiştim?” diye sordurur.
                                         2)) # q the number of moving average terms)) q DÜN NE KADAR SAÇMALADIM
```

```python
results = model.fit()
```

```python
arima_predictions= results.get_forecast(steps=len(testing_set)).predicted_mean
sns.lineplot(training_set)
sns.lineplot(testing_set)
sns.lineplot(arima_predictions)
```

```python
from sklearn.metrics import root_mean_squared_error,mean_absolute_error,mean_squared_error,r2_score
```

```python
def model_evulation(actual_values,predicted_values):
    print(f"RMSE: {np.round(root_mean_squared_error(actual_values, predicted_values),2)}")
    print(f"MAE: {np.round(mean_absolute_error(actual_values, predicted_values),2)}")
    print(f"MSE: {np.round(mean_squared_error(actual_values, predicted_values),2)}")
    print(f"R2: {np.round(r2_score(actual_values, predicted_values),2)}")
```

```python
model_evulation(testing_set,arima_predictions)
```

```python
baseline_ymean = [training_set.mean() for i in range(len(testing_set))]
```

```python
#let's make another baseline: extend the final term indefinitely
# en son degeri baseline degeri olarak aliyo
baseline_last_y = [training_set.iloc[-1] for i in range(len(testing_set))]
```

```python
model_evulation(testing_set, baseline_last_y)
```

```python
# Exponential Smoothing model (ETS)
# Düne yakın olan veri daha önemli, eski veri daha az önemli.
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
ets_models = ExponentialSmoothing(
    endog = training_set,# tahmin etmek istediğin SERİ
    seasonal = 'add', # Bu seride tekrar eden bir desen var ve bu desen veriye ekleniyor.”
    seasonal_periods= 24 )
```

```python
ets_result = ets_models.fit()
```

```python
ets_predictions = ets_result.forecast(steps=len(testing_set))
```

```python
sns.lineplot(training_set[8000:])
sns.lineplot(testing_set)
sns.lineplot( x=testing_set.index,y=ets_predictions)
plt.ylim(0,50)
```

```python
model_evulation(testing_set, ets_predictions)
```
